```python
def get_data_by_threads(service, threads):
    X = []
    Ylat = []
    Yrps = []
    for item in data:
        if item['test_name'].startswith(service) and int(item['threads']) == threads:
            X.append(int(item['connections']))
            Ylat.append(item['latency'])
            Yrps.append(float(item['requests_per_second']))
    
    Ylat_avg = [y['avg'] for y in Ylat]
    stddevs = [y['stddev'] for y in Ylat]
    Ylat_ub = [avg + std for avg, std in zip(Ylat_avg, stddevs)]
    Ylat_lb = [max(0.0,avg - std) for avg, std in zip(Ylat_avg, stddevs)]

    return X,Yrps,Ylat_avg,Ylat_lb,Ylat_ub
```
