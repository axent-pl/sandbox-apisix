```python
def get_data_by_threads(service, threads):
    X = []
    Ylat = []
    Yrps = []
    for item in data:
        if item['test_name'].startswith(service) and int(item['threads']) == threads:
            X.append(int(item['connections']))
            Ylat.append(item['latency'])
            Yrps.append(float(item['requests_per_second']))
    
    Ylat_avg = [y['avg'] for y in Ylat]
    stddevs = [y['stddev'] for y in Ylat]
    Ylat_ub = [avg + std for avg, std in zip(Ylat_avg, stddevs)]
    Ylat_lb = [max(0.0,avg - std) for avg, std in zip(Ylat_avg, stddevs)]

    return X,Yrps,Ylat_avg,Ylat_lb,Ylat_ub
```

```python
import matplotlib.pyplot as plt

service = 'gateway'
threads = [1,2,3,4]


# Plotting
fig, ax = plt.subplots(nrows=len(threads), ncols=2, figsize=(16, 9), squeeze=False, sharex='col', sharey='col')
fig.suptitle(f'{service} tests')
for idx,t in enumerate(threads):
    X,Yrps,Ylat_avg,Ylat_lb,Ylat_ub = get_data_by_threads(service=service, threads=t)

    ax[idx,0].plot(X, Ylat_avg, '-', label=f'Threads={t}', color='blue')
    ax[idx,0].fill_between(X, Ylat_lb, Ylat_ub, color='blue', alpha=0.1)
    ax[idx,0].legend()

    ax[idx,1].plot(X, Yrps, '-', label='Max', color='blue')

    if idx == 0:
        ax[idx,0].set_title('Latency')
        ax[idx,1].set_title('Requests per second')
    if idx == len(threads) - 1:
        ax[idx,0].set(xlabel='Connections', ylabel='Latency [ms]')
        ax[idx,1].set(xlabel='Connections', ylabel='RPS [1]')
    else:
        ax[idx,0].set(ylabel='Latency [ms]')
        ax[idx,1].set(ylabel='RPS [1]')
```
